# Direct Marketing Campaigns of Portuguese Banking Clients
2 years worth of data (2008-2010).
4,119 records.
20 independent variables.  
1 dependent variable.

**Note from the authors of the dataset:**
> Missing Attribute Values: There are several missing values in some categorical attributes, all coded with the "unknown" label. These missing values can be treated as a possible class label or using deletion or imputation techniques.

Will define the NA value when reading in the dataset.

```python
banking_raw = pd.read_csv('./data/bank-original.csv', header=0, sep=";", na_values="unknown")
banking_raw.shape
```

```python
banking_raw.head().T
```

```python
banking_raw.describe(include="all")
```

*Please note, according to the dataset description, a client that had not been contacted previously is represented with the value 999 in the "contact" column.*

```python
banking_raw.isna().sum()
```

Will drop all records that contain any null values since we need to reduce our dataset to <= 3,000 records.

```python
banking_nonulls = banking_raw.dropna(how='any')
banking_nonulls.shape
```

```python
bank = banking_nonulls
```

In summary, we:
- imported our original 4,119 record dataset
- defined & identified the null values present
- removed the rows with any null value present

We will keep the dataset as is (3,090) as permitted by the professor, and will export `banking_nonulls` dataframe to a CSV.

```python
bank.rename(columns={"emp.var.rate": "emp_var_rate", "cons.price.idx": "cons_price_idx", "cons.conf.idx": "cons_conf_idx", "nr.employed": "nr_employed"}, 
            inplace=True)
```

```python
bank.columns
```

```python
bank.reset_index(inplace=True, drop=True)
```

```python
bank.head(10)
```

```python
bank.to_csv('./data/cleaned-bank.csv')
```
