Will define the NA value when reading in the dataset.

```python
banking = pd.read_csv('./data/bank-additional.csv', header=0, sep=";", na_values="unknown")
banking.shape
```

```python
banking.head().T
```

```python
banking.describe(include="all")
```

*Please note, according to the dataset description, a client that had not been contacted previously is represented with the value 999 in the "contact" column.*

```python
banking.isna().sum()
```

Will drop all records that contain any null values since we need to reduce our dataset to <= 3,000 records.

```python
banking_nonulls = banking.dropna(how='any')
banking_nonulls.shape
```

In summary, we:
- imported our original 4,119 record dataset
- defined & identified the null values present
- removed the rows with any null value present

This has lead to a reduction of 803 records.  I recommend we either keep the dataset as is (3,090) or remove 90 more records in order to meet the recommended dataset size.  If we decide to keep dataset as is, we can export `banking_nonulls` dataframe to a CSV.  If we decide to remove more, please make any more removals as you see fit before exporting 


```python
banking_nonulls.to_csv('./data/cleaned-bank.csv')
```
